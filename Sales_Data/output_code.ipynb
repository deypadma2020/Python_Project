```python
import pandas as pd
import numpy as np

pd.pandas.set_option("display.max_columns", None)
```

```python
order_report = pd.read_excel("/Company X - Order Report.xlsx")
order_report.head(2)
```

```python
pincode_zones = pd.read_excel("/Company X - Pincode Zones.xlsx")
pincode_zones.head(2)
```

```python
sku_master = pd.read_excel("/Company X - SKU Master.xlsx")
sku_master.head(2)
```

```python
invoice = pd.read_excel("/Courier Company - Invoice.xlsx")
invoice.head(2)
```

```python
rates = pd.read_excel("/Courier Company - Rates.xlsx")
rates.head(2)
```

```python
rates['Zone'].replace({'A': 'a','B': 'b','C':'c','D':'d','E':'e'}, inplace=True)
rates.head(2)
```

```python
order_report.rename(columns={'ExternOrderNo':'Order ID'}, inplace = True)
order_report.head(2)
```

```python
merged_data1 = pd.merge(order_report,sku_master,on='SKU', how='left')
merged_data1.head(2)
```

```python
merged_data2 = pd.merge(merged_data1,invoice,on='Order ID', how='left')
merged_data2.head(2)
```

```python
merged_data = pd.merge(merged_data2,rates,on='Zone', how='left')
merged_data.head(2)
```

```python
merged_data.shape
```

```python
# Calculate total weight for each order
merged_data['Total Weight (g)'] = (merged_data['Weight (g)'] * merged_data['Order Qty'])
merged_data.head(2)
```

```python
# Determine applicable weight based on the zone's weight slab
merged_data['Applicable Weight (g)'] = merged_data['Charged Weight'] * merged_data['Weight Slabs']

merged_data.head(2)
```

```python
# Calculate Expected Charge based on shipment type
merged_data.loc[merged_data['Type of Shipment'] == 'Forward charges', 'Expected Charge'] = merged_data['Forward Fixed Charge'] + merged_data['Applicable Weight (g)'] * merged_data['Forward Additional Weight Slab Charge']

merged_data.loc[merged_data['Type of Shipment'] == 'Forward and RTO charges', 'Expected Charge'] = merged_data['Forward Fixed Charge'] + (merged_data['Applicable Weight (g)'] * merged_data['Forward Additional Weight Slab Charge']) + (merged_data['Applicable Weight (g)'] * merged_data['RTO Additional Weight Slab Charge'])

merged_data.head(2)
```

```python
# Convert total weight into KG
merged_data['Total Weight (KG)'] = (merged_data['Total Weight (g)'] / 1000)
merged_data.head(2)
```

```python
#columns = list(merged_data.columns)
#columns
```

```python
# Create Output Data 1
resultant_table = merged_data[['Order ID', 'AWB Code', 'Total Weight (KG)', 'Weight Slabs',
                              'Charged Weight', 'Forward Additional Weight Slab Charge',
                              'Zone', 'Zone', 'Expected Charge', 'Billing Amount (Rs.)']]

# Rename columns for clarity
resultant_table.columns = ['Order ID', 'AWB Number', 'Total weight as per X (KG)', 'Weight slab as per X (KG)',
                          'Total weight as per Courier Company (KG)', 'Weight slab charged by Courier Company (KG)',
                          'Delivery Zone as per X', 'Delivery Zone charged by Courier Company',
                          'Expected Charge as per X (Rs.)', 'Charges Billed by Courier Company (Rs.)']

# Calculate Difference
resultant_table['Difference'] = resultant_table['Expected Charge as per X (Rs.)'] - resultant_table['Charges Billed by Courier Company (Rs.)']

resultant_table.head(2)
```

```python
summary_table = pd.DataFrame({
    'Total Orders - Correctly Charged': [len(resultant_table[resultant_table['Difference'] == 0])],
    'Total Orders - Over Charged': [len(resultant_table[resultant_table['Difference'] > 0])],
    'Total Orders - Under Charged': [len(resultant_table[resultant_table['Difference'] < 0])],
    'Total Amount - Correctly Charged': [resultant_table[resultant_table['Difference'] == 0]['Charges Billed by Courier Company (Rs.)'].sum()],
    'Total Amount - Over Charged': [resultant_table[resultant_table['Difference'] > 0]['Charges Billed by Courier Company (Rs.)'].sum()],
    'Total Amount - Under Charged': [resultant_table[resultant_table['Difference'] < 0]['Charges Billed by Courier Company (Rs.)'].sum()]
})
summary_table.T
```

```python
# Create a new dataframe
summary_table = pd.DataFrame({
    'Category': ['Total Orders - Correctly Charged', 'Total Orders - Over Charged', 'Total Orders - Under Charged'],
    'Count': summary_table.iloc[:, :3].values.flatten(),
    'Amount': summary_table.iloc[:, 3:].values.flatten()
})
summary_table
```

```python
with pd.ExcelWriter('assignment.xlsx') as writer:
    summary_table.to_excel(writer, sheet_name='summary',index = False)
    resultant_table.to_excel(writer, sheet_name='calculation', index = False)
```
